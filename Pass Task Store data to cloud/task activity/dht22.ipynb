### Connect to the database

```python
import firebase_admin
from firebase_admin import credentials

databaseURL = "https://dht22-1e9d4-default-rtdb.asia-southeast1.firebasedatabase.app/"

cred = credentials.Certificate("dht_credentials.json")
firebase_admin.initialize_app(cred,
                              {'databaseURL':databaseURL})
```

### Write to the database

```python
# import data
import pandas as pd
df = pd.read_csv("dht22 data.csv", header=None, names=['Timestamp', 'Humidity', 'Temperature'])
df.head(), df.columns
```

```python
# convert dataframe to json
json_data = df.to_json(orient='index', date_format='iso')

# save the json file
with open('dht22_data.json', 'w') as json_file:
    json_file.write(json_data)
```

```python
from firebase_admin import db
import json

# reference to the root of the DB
ref = db.reference("/") 

# overwrite the the reference point at root node
with open("dht22_data.json", "r") as f:
	file_contents = json.load(f)
ref.set(file_contents)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# subplot
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# x accelerometer
accelerometer_x = pd.read_csv('Tom_s realme Thing-Accelerometer_X.csv')
accelerometer_x['time'] = pd.to_datetime(accelerometer_x['time'])
sns.lineplot(data=accelerometer_x, x='time', y='value', ax=axes[0, 0])
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].set_title('X Accelerometer')

# y accelerometer
accelerometer_y = pd.read_csv('Tom_s realme Thing-Accelerometer_Y.csv')
accelerometer_y['time'] = pd.to_datetime(accelerometer_y['time'])
sns.lineplot(data=accelerometer_y, x='time', y='value', ax=axes[0, 1])
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].set_title('Y Accelerometer')

# z accelerometer
accelerometer_z = pd.read_csv('Tom_s realme Thing-Accelerometer_Z.csv')
accelerometer_z['time'] = pd.to_datetime(accelerometer_z['time'])
sns.lineplot(data=accelerometer_z, x='time', y='value', ax=axes[1, 0])
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].set_title('Z Accelerometer')

# Combine accelerometer
sns.lineplot(data=accelerometer_x, x='time', y='value', ax=axes[1, 1])
sns.lineplot(data=accelerometer_y, x='time', y='value', ax=axes[1, 1])
sns.lineplot(data=accelerometer_z, x='time', y='value', ax=axes[1, 1])
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].set_title('Accelerometer')

plt.tight_layout()
plt.show()
```

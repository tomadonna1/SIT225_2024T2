```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('8_2_data.csv')
df['Timestamp'] = pd.to_datetime(df['Timestamp']).dt.strftime('%H')
df.head()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

sns.lineplot(data=df, x='Timestamp', y='accelerometer_X', ax=axes[0, 0])
axes[0, 0].tick_params(axis='x', rotation=45)
axes[0, 0].set_title('X Accelerometer')

sns.lineplot(data=df, x='Timestamp', y='accelerometer_Y', ax=axes[0, 1])
axes[0, 1].tick_params(axis='x', rotation=45)
axes[0, 1].set_title('Y Accelerometer')

sns.lineplot(data=df, x='Timestamp', y='accelerometer_Z', ax=axes[1, 0])
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].set_title('Z Accelerometer')

sns.lineplot(data=df, x='Timestamp', y='accelerometer_X', ax=axes[1, 1])
sns.lineplot(data=df, x='Timestamp', y='accelerometer_Y', ax=axes[1, 1])
sns.lineplot(data=df, x='Timestamp', y='accelerometer_Z', ax=axes[1, 1])
axes[1, 1].tick_params(axis='x', rotation=45)
axes[1, 1].set_title('Accelerometer')

plt.tight_layout()
plt.show()
```
